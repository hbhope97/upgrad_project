Checking Missing Values

```python
for col in df.columns:
  x = df[col].isna().sum()*100/df[col].shape[0]
  print(col +"         " , x)
```

In the above, we focus on columns with very high NAN values, which are mostly related to recharge data or services(>70% NA values). The implication of these columns is that the customer has probably stopped using the service, which is why it is showing up as NaN. We will be looking at grouping togethers columns for each month.

```python
junecols = ['night_pck_user_6', 'arpu_2g_6', 'count_rech_3g_6',
       'av_rech_amt_data_6', 'count_rech_2g_6', 'max_rech_data_6', 'fb_user_6',
       'date_of_last_rech_data_6', 'total_rech_data_6', 'arpu_3g_6']
julycols = ['max_rech_data_7', 'av_rech_amt_data_7', 'date_of_last_rech_data_7',
       'count_rech_3g_7', 'fb_user_7', 'arpu_2g_7', 'total_rech_data_7',
       'count_rech_2g_7', 'arpu_3g_7', 'night_pck_user_7']
augcols = ['arpu_2g_8', 'date_of_last_rech_data_8', 'arpu_3g_8',
       'night_pck_user_8', 'fb_user_8', 'av_rech_amt_data_8',
       'count_rech_3g_8', 'count_rech_2g_8', 'max_rech_data_8',
       'total_rech_data_8']
sepcols = ['date_of_last_rech_data_9', 'total_rech_data_9', 'count_rech_3g_9',
       'arpu_3g_9', 'count_rech_2g_9', 'arpu_2g_9', 'max_rech_data_9',
       'night_pck_user_9', 'av_rech_amt_data_9', 'fb_user_9']
```

```python
for col in junecols+julycols+augcols+sepcols:
  df[col].fillna(0, inplace=True)
```

We imputed the columns with zero becasue the assumption is that they are all 0 because the customer stopped using those services.

```python
#recheck missing value:

for col in df.columns:
  x = df[col].isna().sum()*100/df[col].shape[0]
  print(col +"         " , x)
```

We can see that that there's still some missing values left, in this case we will use the date_of_last_rech as an indicator that the customer stopped all the services for the month, and impute the corresponding columns for that month with zero.

```python
replace_array = ['date_of_last_rech_6','date_of_last_rech_7','date_of_last_rech_8','date_of_last_rech_9']
for col in replace_array:
  df[col].fillna(0, inplace=True)
```

```python
june_col = [col for col in df.columns
                    if df[col].isnull().sum() > 0 and col.endswith('6')]
july_col = [col for col in df.columns
                    if df[col].isnull().sum() > 0 and col.endswith('7')]
aug_col = [col for col in df.columns
                    if df[col].isnull().sum() > 0 and col.endswith('8')]
sep_col = [col for col in df.columns
                    if df[col].isnull().sum() > 0 and col.endswith('9')]
print(sep_col)
```

```python
for col in june_col:
    df.loc[df["date_of_last_rech_6"] == 0, col] = 0
for col in july_col:
    df.loc[df["date_of_last_rech_7"] == 0, col] = 0
for col in aug_col:
    df.loc[df["date_of_last_rech_8"] == 0, col] = 0
for col in sep_col:
    df.loc[df["date_of_last_rech_9"] == 0, col] = 0
```

As the customer stopped using the services we also have to impute the columns of services with zero whenever the corresponding value in date_of_last_reach is 0.

```python
#check missing values again
for col in df.columns:
  x = df[col].isna().sum()*100/df[col].shape[0]
  print(col +"         " , x)
```

We notice that there are still missing values in the services columns for each month. This indicates that the customer paid for the services but did not use them. We will impute these values with zero.

```python
for col in june_col:
  df[col].fillna(0, inplace=True)
for col in july_col:
  df[col].fillna(0, inplace=True)
for col in aug_col:
  df[col].fillna(0, inplace=True)
for col in sep_col:
  df[col].fillna(0, inplace=True)
```

```python
for col in df.columns:
  print(df[col].isna().sum())
```

From the above data, we will target filter top customers, i.e customers above a certain percentile in recharge amount

```python
df["data_recharge_6"] = df["total_rech_data_6"].copy() * df["av_rech_amt_data_6"].copy()
df["data_recharge_7"] = df["total_rech_data_7"].copy() * df["av_rech_amt_data_7"].copy()
```

```python
df['total_avg_rech_amt_good_phase'] = (df["total_rech_amt_6"].copy() + df["total_rech_amt_7"].copy() +
                                       df["data_recharge_6"].copy() + df["data_recharge_7"].copy())/2
tc = df[df["total_avg_rech_amt_good_phase"] >= df["total_avg_rech_amt_good_phase"].quantile(0.7)]
```

We will define churned customers for September, i.e customers who have not made any incoming, outgoing calls or used mobile internet

```python
#defining new dataframe because of mem limits

churn = pd.DataFrame(index = tc.index)

churn.loc[:,'churn'] = tc.loc[:,"total_ic_mou_9"].copy() + tc.loc[:,"total_og_mou_9"].copy() + tc.loc[:,"vol_2g_mb_9"].copy() + tc.loc[:,"vol_3g_mb_9"].copy()
churn.loc[:,"churn"] = churn.loc[:,"churn"].copy().apply(lambda x: 1 if x==0 else 0) # 1 is churned customer, 0 is non-churned customer
```

```python
#remove all september columns from the dataframe:
columns_to_drop = df.columns[tc.columns.str.endswith("9")]
tc.drop(columns_to_drop, axis=1, inplace = True)

#drop the vbg column
tc.drop("sep_vbc_3g", axis = 1, inplace = True)

print(tc)
```

```python
#changing values of date of last-recharge to indicate if the user recharged or not.

date_type = ['date_of_last_rech_6', 'date_of_last_rech_7', 'date_of_last_rech_8',
             'date_of_last_rech_data_6', 'date_of_last_rech_data_7','date_of_last_rech_data_8']
for col in date_type:
    tc[col] = tc[col].apply(lambda x: 0 if x==0 else 1)
```

Modeling with XGBoost

```python
y = churn
X = tc
```

```python
y.value_counts(normalize=True)*100
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score
```

The total number of churns is much lower than nochurn. However, XGBOOST can handle imbalanced data with class-weights

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
model = xgb.XGBClassifier(learning_rate = 0.3, subsample = 0.5, scale_pos_weight=91.86) #(this defines the class imbalance ratio )
model.fit(X_train, y_train)
print(classification_report(y_test, model.predict(X_test)))
print("ROC score:", roc_auc_score(y_test, model.predict(X_test)))
```

Clealy the model can predict non-churners well, but we want to improve the recall rate on churners. The precision rate is not as important since identifying non-churners as churners is not that problematic

Feature Importance on xgboost

We start with the simplest one, the one which is provided with xgboost.

```python
feature_importances = model.get_booster().get_score(importance_type='gain')

print(feature_importances['aon'])
print(feature_importances['jun_vbc_3g'])
print(feature_importances['jul_vbc_3g'])
print(feature_importances['aug_vbc_3g'])
print(feature_importances['total_avg_rech_amt_good_phase'])

del feature_importances['aon']
del feature_importances['jun_vbc_3g']
del feature_importances['jul_vbc_3g']
del feature_importances['aug_vbc_3g']
del feature_importances['total_avg_rech_amt_good_phase']

A = set()

for key in feature_importances.keys():
  if key.endswith('_6') or key.endswith('_7') or key.endswith('_8'):
    A.add(key[:-2])

data = pd.DataFrame(index = list(A), columns = ["June","July","August"])

for x in data.index:
  try:
    data.loc[x,"June"] = feature_importances[x + "_6"]
  except KeyError:
    data.loc[x,"June"] = 0
  try:
    data.loc[x,"July"] = feature_importances[x + "_7"]
  except KeyError:
    data.loc[x,"July"] = 0
  try:
    data.loc[x,"August"] = feature_importances[x + "_8"]
  except KeyError:
    data.loc[x,"August"] = 0


data["mean importance june/july"] = data.loc[:,["June",'July']].mean(axis = 1)

print(data["August"].sort_values())
```

The most important features for the month of august are:
*   Minutes of outgoing roaming
*   Minutes of ISD outgoing
*   Minutes of special calls incoming
*   Incoming calls from other operator that are local

```python
print(data['mean importance june/july'].sort_values())
```

The most important features for the 'good' months are:
*   Usage of low-validity 3g sachets
*   Local incoming from other operators
*   Total incoming minutes
*   local outgoing to other operators
